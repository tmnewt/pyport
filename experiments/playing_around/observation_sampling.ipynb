Goal of this notebook is to solve the problem of turning daily data into lower frequencies. That is, take daily data for say, the opening prices of facebook and find the weekly opening price. Do the same for each of the prices. Maybe also throw in volumn there for good measure...


```python
# let's get some data to play with.
import time
import pandas as pd
import pandas_datareader as pdr
from findatapy.market import Market, MarketDataGenerator, MarketDataRequest
from findatapy.util import SwimPool; SwimPool()

market = Market(market_data_generator=MarketDataGenerator())

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
data_request = MarketDataRequest(
    data_source='yahoo',
    start_date='year',
    tickers=['FB', 'GOOGL'],
    fields=['open', 'close', 'high','low'],
    vendor_tickers=['FB','GOOGL'],
    vendor_fields=['Open','Close', 'High','Low'])
df = market.fetch_market(data_request)
df.tail()
```

Compare tha with the normal pandas data reader.

```python
ticker_list = ['LOW','BKNG','DD','AMT','GE','RSG','SNPS','VEEV','CBS','DXC','BG','LAMR','KIM','HELE','SIGI','S','FHB','CRS','AIMC','TEX','FORM','CSGS','AROC','ATRA','ANDE','TELL','GFF','LMNX','ACLS','GOOD','AKBA','RILY','TLRA','LAND','FBMS','NKSH','TWIN']

df_pdr = pdr.get_data_yahoo(ticker_list, '11/25/09', '11/26/19', interval = 'd')
```

```python
pd.set_option('display.max_columns',250)
df_pdr.resample('W').ohlc()
```

Recall that for findata, the dataframe comes back without multi-indexing

```python
data_request = MarketDataRequest(
    data_source='yahoo',
    start_date='decade',
    tickers=ticker_list,
    fields=['open', 'close', 'high','low', 'adjclose'],
    vendor_tickers=ticker_list,
    vendor_fields=['Open','Close', 'High','Low','Adj Close'])
df = market.fetch_market(data_request)
df.tail()
```

It's probably a good idea to figure out how to sample from either dataframe type. Never know when I might need both.

Wait, I know, pandas has nice method `ohlc()` which generates the 'open', 'high', 'low', and 'close' for the resample. For instance, let's resample the past year of facebook's prices so that the observations are in weeks instead of days.

```python
data_year_fb = MarketDataRequest(
    data_source='yahoo',
    start_date='year',
    tickers = ['Facebook'],
    fields = ['open','high','low','close'],
    vendor_tickers=['FB'],
    vendor_fields=['Open','High','Low', 'Close'])
df_year_fb = market.fetch_market(data_year_fb)
```

```python
# normal look
df_year_fb.head()
df_year_fb.resample('w').ohlc().head()
```
